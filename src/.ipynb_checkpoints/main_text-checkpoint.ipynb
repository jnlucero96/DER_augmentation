```python
def compute_LCOPR(OCS, PV2GRID, PR, N, E0):
    
    numerator = OCS - PV2GRID*N*PR
    denominator = N*E0

    if E0 == 0.0:
        return 0.0
    else:
        return numerator/denominator
```

```python
def compute_LCOS(OCS, E_rated, N_EOL):

    numerator = OCS
    denominator = N_EOL*E_rated

    return numerator/denominator
```

# Figure 6: Total electricity cost (TEC) for viable solar sites in the Port of Savannah region.

```python
fig, ax = subplots(figsize=(6, 3.5))

NMC_100MWh_cost = batt_LCOS[0, -1]
NCA_100MWh_cost = batt_LCOS[1, -1]
LFP_100MWh_cost = batt_LCOS[2, -1]

cost_no_batt = fractions_optimal[sorted_df_reduced_gid, 1, 0, 0]*site_LCOPR_woReturns[:,1] + fractions_optimal[sorted_df_reduced_gid, 1, 0, 2]*160
cost_w_batt_NMC = (
    fractions_optimal[sorted_df_reduced_gid, 1, 1, 0]*site_LCOPR_woReturns[:,1] 
    + fractions_optimal[sorted_df_reduced_gid, 1, 1, 1]*(fractions_batt_charge_optimal[sorted_df_reduced_gid, 1, 1, 0]*site_LCOPR_woReturns[:,1] + fractions_batt_charge_optimal[sorted_df_reduced_gid, 1, 1, 1]*160 + NMC_100MWh_cost) 
    + fractions_optimal[sorted_df_reduced_gid, 1, 1, 2]*160
    )
cost_w_batt_NCA = (
    fractions_optimal[sorted_df_reduced_gid, 1, 1, 0]*site_LCOPR_woReturns[:,1] 
    + fractions_optimal[sorted_df_reduced_gid, 1, 1, 1]*(fractions_batt_charge_optimal[sorted_df_reduced_gid, 1, 1, 0]*site_LCOPR_woReturns[:,1] + fractions_batt_charge_optimal[sorted_df_reduced_gid, 1, 1, 1]*160 + NCA_100MWh_cost) 
    + fractions_optimal[sorted_df_reduced_gid, 1, 1, 2]*160
    )
cost_w_batt_LFP = (
    fractions_optimal[sorted_df_reduced_gid, 1, 1, 0]*site_LCOPR_woReturns[:,1] 
    + fractions_optimal[sorted_df_reduced_gid, 1, 1, 1]*(fractions_batt_charge_optimal[sorted_df_reduced_gid, 1, 1, 0]*site_LCOPR_woReturns[:,1] + fractions_batt_charge_optimal[sorted_df_reduced_gid, 1, 1, 1]*160 + LFP_100MWh_cost) 
    + fractions_optimal[sorted_df_reduced_gid, 1, 1, 2]*160
    )

ax.axhline(160, ls='--', color='r', zorder=1, label=r"$\mathrm{Only\ grid}$")
ax.plot(arange(171) + 1, cost_no_batt, "b-", ms=6, label=r"$\mathrm{No\ Battery}$")
ax.plot(arange(171) + 1, cost_w_batt_NMC, "C0-", ms=6, label=r"$\mathrm{NMC}$")
ax.plot(arange(171) + 1, cost_w_batt_NCA, "C1-", ms=6, label=r"$\mathrm{NCA}$")
ax.plot(arange(171) + 1, cost_w_batt_LFP, "C2-", ms=6, label=r"$\mathrm{LFP}$")

handles_og, labels_og = ax.get_legend_handles_labels()
handles1, labels1 = handles_og[:2], labels_og[:2]
handles2, labels2 = handles_og[2:], labels_og[2:]

leg0 = ax.legend(
    handles1, labels1, 
    loc="center", prop=dict(size=12), ncol=2, 
    bbox_to_anchor=(0.5, 1.1), 
    bbox_transform=ax.transAxes
    )

ax.legend(
    handles2, labels2, 
    loc="lower left", prop=dict(size=12), 
    title=r"$\mathrm{Battery\ chemistry}$",
    title_fontsize=14, ncol=3
    )
ax.add_artist(leg0)

ax.set_ylim((0, None))
ax.set_xlim((0, 173))

ax.set_ylabel(r"$\mathrm{TEC\ (\$/MWh)}$")
ax.set_xlabel(r"$\mathrm{Site\ index}$")

def val2per(x):
    return 1e2*x/160

def per2val(x):
    return (160*x/1e2)

secax = ax.secondary_yaxis('right', functions=(val2per, per2val))
secax.set_ylabel(r'$\mathrm{Percent\ of\ baseline\ (\%)}$')

fig.tight_layout()
fig.align_labels()

fig.subplots_adjust(top=0.87)

fig.savefig("../figures/Fig6.pdf")
```
